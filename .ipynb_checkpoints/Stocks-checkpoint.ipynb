# Loading in the data

Below I loaded in the data three different times to make data frames for AAPL, MSFT, IBM. 

## Parameters used in read_csv()

--------------------------------------------------------------------------------
### ```header = 1```
When reading the csv file raw the first line **```AAPL,,,,,MSFT,,,,,IBM,,,,```** gets parsed as the names of columns. 

Given that these titles are only three strings, they can be remebered as titles and used as the **names in assignment**. Therefore, with the code **``` header = 1 ```** this line can be excluded completely using the next line as the columns.

--------------------------------------------------------------------------------

### ```usecols = [0,1,2,3,4,...]```
Calling columns by their name picks up only the first instance of that name. Meaning if I used ```usecols = [Date, High, Low,...]``` I would be calling the columns **only** for AAPL and not from the ones after. 

Instead I opted to use the **indexes** of columns as the identifiers. There are three copies of each column making 3 sets of 5 columns. 
- **AAPL** -  [0, 1, 2, 3, 4]
- **MSFT** -  [5, 6, 7, 8, 9]
- **IBM** -  [10, 11, 12, 13, 14]

--------------------------------------------------------------------------------
### ```names = ['date','high','low','close','volume']```

This parameter was used to set standardized column names, which in this case just decapitalized the first letter in each name.  

--------------------------------------------------------------------------------
### ```parse_dates = ['date']```

Each observation for the column date was converted to a datetime object, rather than a string.  

--------------------------------------------------------------------------------
### ```index_col = 'date'```

Sets the index column to be the values of the date column, meaning that the first couple of rows will look like so:
```
	          high	  low	    close	  volume
date				
1980-12-12	0.5156	0.5134	0.5134	117258400.0
1980-12-15	0.4888	0.4866	0.4866	43971200.0
1980-12-16	0.4531	0.4509	0.4509	26432000.0
```

--------------------------------------------------------------------------------
### Removing ```NaN``` with ```.dropna()```

The three stocks went public at different times, meaning that there are different amount of rows to read in for each stock. The function ```.dropna()``` removes of any ```NaN``` that may be at the end of the tables for the newer stocks. 


#### AAPL
This section contains the assigned dataframe for AAPL

````python
import pandas as pd 

# Assign AAPL dataframe to corresponding title
aapl = pd.read_csv('rawdata.txt', # source file 
                 header=1,    # Gets rid of the first line
                 usecols= [0,1,2,3,4],    # Uses first five indexed columns
                 names = ['date','high','low','close','volume'],    # renames selected columns
                 parse_dates=['date'],     # Parses date to make datetime type
                 index_col= 'date'      # date column as an index
                 ).dropna()     # Removes NaN 
````

To give some colour to the information in the table is the metadata. From we can see that there are no **non-null entries** in table and entries are **equivalent** across all variables.  

It can also be observed that all variables have been read in as **floats** (float64). This is acceptable for varibles ```high, low, and close``` but since ```volume``` is the number of shares sold, this technically would be a discrete variable and so potentially **should be** read in as an **integer**. 

   

````python
aapl.info()
````

The chunk below tests whether there are observations in the column ```volume``` that do not end in a ```.0```. Since all of the volume column are in fact intergers they are also converted to the data type 'int'. 

Changes can be seen in the ```.info()``` output.

````python
# Tests if volume column only contains integers and converts it to an integer if so
aapl = (aapl.astype({'volume':'int'}) if len([i for i in list(aapl['volume']) if i % 1 !=0])==0 else aapl)

# Changes can be seen below
aapl.info()
````

````python
aapl.head()
````

# Closing prices over time

The plot below shows the closing times of ```AAPL, MSFT,& IBM``` over 1980-2017. Here are some of the visual trends:

####**Rises and falls during the millenium shift** 
- *Rises*
  - IBM & MSFT saw substantial growth arround 1999
  - AAPL saw relatively large growth in 2000
- *Falls*
  - IBM was volatile but held till around 2003 when it plummeted
  - MSFT held their growth for a shorter period and dropped dramatically around 2000
  - AAPL dropped suddenly in around 2001

####**Growth in the 2010s**
- **AAPL** have trended upwards, with some volatilty, making them **valued highest** out of the three stocks by 2017
- **MSFT** saw really consistent and stable growth
- **IBM** had dramatic growth at the start of the decade, but their fall in 2015 is still being recovered from 


````python
#@title ___
from matplotlib import pyplot as plt

# Setting plot size
plt.figure(figsize=(12,5))

# Set plot style
plt.style.use('ggplot')

# Set plots for each stock with labels
plt.plot(aapl.index,aapl.close, label = 'AAPL', lw = 1.2)     # Line plot
plt.fill_between(aapl.index,aapl.close, alpha=0.3)  # Area plot with tranparency

plt.plot(msft.index,msft.close, label ='MSFT', lw = 1.2)      # Line plot
plt.fill_between(msft.index,msft.close, alpha=0.3)  # Area plot with tranparency

plt.plot(ibm.index,ibm.close, label = 'IBM', lw = 1.2)        # Line plot
plt.fill_between(ibm.index,ibm.close, alpha=0.2)    # Area plot with tranparency

# Setting tick intervals for y axis
plt.yticks(range(0,250,25))

# Set axis labels
plt.xlabel('Year')
plt.ylabel('Closing Share Price')

# Display legend
plt.legend()

# Remove gridlines
plt.grid(False)

# Add a title
plt.title('Closing prices of AAPL, MSFT & IBM')
````

# Measuring Fluctuation 
**Constructed a *fluctuation* metric to measure the range of the stock value that day as a percentage of the closing value.** 

````python
aapl['fluctuation'] = 100*((aapl['high'] - aapl['low'])/aapl['close'])

msft['fluctuation'] = 100*((msft['high'] -msft['low'])/msft['close'])

ibm['fluctuation'] = 100*((ibm['high'] - ibm['low'])/ibm['close'])
````

# Fluctuation over time
**Plots below describe the distribution of fluctuation alongside the closing share price over the range of the collected data (1980-2017).**

````python
#@title AAPL
from matplotlib import pyplot as plt

# Set style of the Plot
plt.style.use('ggplot')

# Creates subplots assigned to fig and ax1 
fig, ax1 = plt.subplots(figsize=(12, 5))

# Duplicates ax1 to make the 2nd y-axis ax2
ax2 = ax1.twinx()


# Plot of closing times 
ax1.plot(aapl.index,aapl.close,     
         lw = 1,                  # Line width
         alpha = 0.5,             # Set transparency to enable comparisons
         color = 'navy',          # Color
         label = 'Price')         # Legend label


# Plot of fluctuations
ax2.plot(aapl.index,aapl.fluctuation,
            alpha = 0.4,                  # Set transparency to enable comparisons
            color = 'purple',             # Color
            label = 'Fluctuation (%)')    # Legend label

# Axis labels
ax1.set_ylabel('Closing Share Price')   # Label of 1st y-axis  
ax1.set_xlabel('Year')    # Label of x-axis 
ax2.set_ylabel('Fluctuation (% of Close)') # Label of 2nd y-axis

# Remove gridlines
ax2.grid(False)
ax1.grid(False)          

# Show legends
plt.legend(loc = 'upper left', facecolor = 'white') 
ax1.legend(facecolor = 'white', loc = 'upper right')

# Adding a title
plt.title('Daily fluctuations of AAPL with daily close values')
````

# The max fluctuations of ```AAPL``` using idxmax()

````python
date = aapl['fluctuation'].idxmax()   # index of maximun fluctuation
fluct = aapl['fluctuation'][date]     # Returns max fluctuation using date variable as row index

f"AAPL's max fluctuation was on {date} and was {fluct} of the closing price." # Formatted statement
````

This day was **Black Monday** - a stock market crash that lead to a worldwide recession.
